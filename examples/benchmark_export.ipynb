```python
import pickle
import os

BENCHMARK_ROOT = "/data/nas-gpu/wang/atong/Datasets/Benchmark"
experiment = "marina-jittering-test-jittering-0-5-seed-1"
results = pickle.load(open(os.path.join(BENCHMARK_ROOT, "benchmarks", f"{experiment}_benchmark_results.pkl"), 'rb'))
```

```python
import pandas as pd
formatted = [
    {
        'npid': v['npid'],
        'dereplication': (
            -1
            if not any(v['predictions']['dereplication_topk'].values()) 
            else next((k for k, v in v['predictions']['dereplication_topk'].items() if v))
        ),
        'cosine_sim': v['predictions']['cosine_sim'].item(),
        'h_nmr': "\n".join(
            "\t".join(f'{item:.3f}' for item in row)
            for row in v['input']['h_nmr'].tolist()
        ),
        'c_nmr': "\n".join(
            "\t".join(f'{item:.3f}' for item in row)
            for row in v['input']['c_nmr'].tolist()
        ),
        'hsqc': "\n".join(
            "\t".join(f'{item:.3f}' for item in row)
            for row in v['input']['hsqc'].tolist()
        ),
        'mw': round(v['input']['mw'], 5),
        'smiles': v['smiles'],
        '\t': '\t'
    }
    for k, v in results.items()
]
df = pd.DataFrame(formatted)
df.to_excel(os.path.join(BENCHMARK_ROOT, "benchmarks", f"{experiment}_benchmark_results.xlsx"), index=False)
```
